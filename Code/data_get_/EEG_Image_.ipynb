```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eeg_data = np.load('../Data_/chb01/label_1.npy')
eeg_data_sample = eeg_data[:,0:1000]
print(eeg_data_sample.shape)
```

```python
plt.figure(figsize=(8, 5))

for i in range(eeg_data_sample.shape[0]):
    plt.plot(eeg_data_sample[i] + i*10, label=f'Channel {i+1}') 

plt.xlabel('Time')
plt.ylabel('Channels')
plt.title('EEG Data Visualization')

plt.show()
```
